<h2>4) Create new features for training dataset</h2>  

In order to properly train a classifier, it is important that the data is correctly labelled. In this work, the objective is to detect the failures 60 days before it occurs and isolate the component of the turbine that is going to fail. Therefore, a “Failure component” column was added, based on Failure dataset. The column takes the following labels/values:

- `Gearbox: 1`
- `Generator: 2`
- `Generator_Bearing: 3`
- `Transformer: 4`    
- `Hydraulic: 5`
- `The incoming failure is after 60 days: -1`

Note: Sometimes samples are missing when the failure occurs, so you have to specify the timestamp immediately preceding it.


```python
# load the failures in the training set and set up the index
Failures_tr=pd.read_csv('failures_tr.csv')
Failures_tr.set_index('Turbine_ID', inplace=True)

# reset the index of training data
Data_tr.reset_index(inplace=True, drop=False)
```

<h4> Define a function to add the component column to each turbine data </h4>

```python
def add_com_variable(Tur_data, time, com):
    
    dys='60 days'     # Change it to timedelta in order to subtract from another timedelta. 
    dys= pd.to_timedelta(dys)    #Timedeltas are absolute differences in times, expressed in difference units
    
    # index in time and component list
    k=0 
    
    # list for component column
    lc=np.empty
    lc = np.empty(len(Tur_data.index))
    lc[:] = -1
    
    for index, row in Tur_data.iterrows():   #Iterate over DataFrame rows
        if k==len(time):
             break
        a=row.Timestamp   # take the timestamp for each row 
        b=pd.to_datetime(time[k])    # take the time from the time list and change it to datetime
        c=b-a  # subtract the two timestamp, the result is timedelta.
        if c<=dys:   # compare the two timedelta, if c<60 days 
            lc[index]=com[k]  # change the content from -1 to the failure label
            
        # increase k to take the next failure    
        for i in time:   
            if row.Timestamp==pd.to_datetime(i):
                k+=1
                
    # Add the created list to the Turbine T01 data and print the moment when the failure occurred.            
    Tur_data.insert(81,"Component",lc, True)
    return Tur_data
```

*Remark*: The function parameters (time and com) must be defined from the oldest to the newest.
