# 3) PCA

```python
def scree_plot(pca):
    PC_values = np.arange(pca.n_components_) + 1
    PC_labels = ['PC' + str(nb+1) for nb in range(pca.n_components_)] 
    plt.figure(figsize=(8,6))
    plt.bar(PC_values, pca.explained_variance_, linewidth=2, edgecolor='k')
    plt.xticks(ticks=PC_values, labels=PC_labels)
    plt.title('Scree Plot')
    plt.xlabel('Principal Components')
    plt.ylabel('Explained Variance ')
    plt.show()

from sklearn.decomposition import PCA
pca = PCA()
uhhh_pca = pca.fit_transform(Xtr) # get the Principal components
```

```python
scree_plot(pca)
```

```python
def pca_summary(pca, X, out=True):
    """Display a table of the explained std, proportion of variance, 
    and proportion of variance ratio for each component
    """
    names = ["PC"+str(i) for i in range(1, len(pca.explained_variance_ratio_)+1)]
    a = list(np.std(pca.transform(X), axis=0, ddof=1))
    b = list(pca.explained_variance_ratio_)
    c = [np.sum(pca.explained_variance_ratio_[:i]) for i in range(1, len(pca.explained_variance_ratio_)+1)]
    columns = pd.MultiIndex.from_tuples([("sdev", "Standard deviation"), ("varprop", "Proportion of Variance"), ("cumprop", "Cumulative Proportion")])
    summary = pd.DataFrame(list(zip(a, b, c)), index=names, columns=columns)
    if out:
        print("Importance of components:")
        display(summary)
    return summary

_ = pca_summary(pca, Xtr)
```

```python
n = 17;
my_pcas = [f'PC{_}' for _ in range(1,n+1)]
pca_components = pd.DataFrame(pca.components_[:n, :].T, columns=my_pcas, index=Xtr.columns)
```

```python
def biplot2D(score,coeff,labels=None):
    """Generate biplot for the first two principal components 
    to display both scores and variables
    """
    
    xs = score[:,0] # projection on PC1
    ys = score[:,1] # projection on PC2
    p = coeff.shape[1]
    n = score.shape[0]
    
    fig, ax = plt.subplots(figsize=(10,8))
    # plot the scores with the index of the sample
    ax.scatter(xs, ys, marker=".", color = 'k')
    for i in range(33):
        ax.text(xs[i], ys[i], str(i), color = 'k')
    ax.set_xlabel("PC{}".format(1))
    ax.set_ylabel("PC{}".format(2))
    
    # plot the variable vectors (arrow) in the PC plane (loadings)  
    arrow_sc = 5
    color = 'tab:red'
    ax2 = ax.twinx() # instantiate a second x axe
    ax2.set_ylim(-1.2,1.2)
    ax2.tick_params(axis='y', labelcolor=color)
    ax2 = ax2.twiny() # instantiate a second y axe
    ax2.set_xlim(-1.2,1.2)
    ax2.tick_params(axis='x', labelcolor=color)
    for i in range(p):
        ax2.arrow(0, 0, coeff[0, i], coeff[1, i], color = color ,alpha = 0.5, 
                  linestyle = '-',linewidth = 0.01, head_width=0.02, head_length=1)
        if labels is None:
            ax2.text(coeff[0, i]* arrow_sc, coeff[1,i] * arrow_sc, "Var"+str(i+1), 
                     color = color, ha = 'center', va = 'center')
        else:
            ax2.text(coeff[0, i]* arrow_sc, coeff[1, i] * arrow_sc, labels[i], 
                     color = color, ha = 'center', va = 'center')

biplot2D(uhhh_pca[:,0:2], pca.components_[0:2, :], Xtr.columns)
plt.show()
```

# 4) General predictive model

With the available dataset, there are various predictive maintenance strategies that could be developed:
 - Strategy 1- **a specific model for each turbine**. Creating a predictive model for each wind turbine may be precise. For each predictive model, the data from 2016 is used for training while the data from 2017 is used for testing. This strategy could lead to good results, since each model will be confined to a specific wind turbine system. However, due to the distribution of failures in the training and testing set, it will be impossible for the models to detect failures that did not happen in the training set. Therefore, this strategy will not be developed in this notebook.
 - Strategy 2- **a general model for all turbines**. Creating a general predictive maintenance model for all the wind turbines allows ML algorithms to train on more than one source of data, which may provide an informative training set, that may lead to a good prediction model. As illustrated by the figure below, the 2016 data for all the turbines are combined into a single training set, while the test sets are the data for each turbine in 2017.

## A - Machine learning algorithms training
<h6> WARNING: Jump to the models import section and import the trained model, because the training takes long time.</h6>


```python
# import the required libraries
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Redefine Xtr (Ytr doesn't need to change, it's just the component column)
Xtr = sensitive_data


## Initializing Classifiers
clf1 = DecisionTreeClassifier(random_state=1)
clf2 = RandomForestClassifier(random_state=1)
clf3 = MLPClassifier(random_state=1, max_iter=500)


## Setting up the parameter grids
param_grid1 = [{'max_depth': list(range(1, 50)) + [None],
                'criterion': ['gini', 'entropy']}]
param_grid2 = [{'n_estimators': [10, 20, 30, 40, 50]}]
n = len(sensitive_data.iloc[1,:])
param_grid3 = [{'activation': ['identity', 'logistic', 'tanh','relu'],
                'hidden_layer_sizes': [n,2*n,3*n,4*n],
               'learning_rate': ['constant', 'invscaling', 'adaptive']}]


# Setting up multiple GridSearchCV objects, 1 for each algorithm
gridcvs = {}
inner_cv = StratifiedKFold(n_splits=2, shuffle=True, random_state=1)

for pgrid, est, name in zip((param_grid1,param_grid2,param_grid3),
                            (clf1,clf2,clf3),
                            ('DTreeCls','RFCls','MLP')):
    gcv = GridSearchCV(estimator=est,
                      param_grid=pgrid,
                      scoring='accuracy',
                      n_jobs=-1,
                      cv=inner_cv,
                      verbose=0,
                      refit=True)
    gridcvs[name] = gcv 
```

```python
from time import time

for name, gs_est in sorted(gridcvs.items()):

    print(50 * '-', '\n')
    print('Algorithm:', name)
    print('    Inner loop:')
    calculation_start_time = time()

    outer_scores = []
    outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)


    for train_idx, valid_idx in outer_cv.split(Xtr,Ytr):
        gridcvs[name].fit(Xtr.iloc[train_idx], Ytr.iloc[train_idx]) # run inner loop hyperparam tuning
        print('\n        Best ACC (avg. of inner test folds) %.2f%%' % (gridcvs[name].best_score_ * 100))
        print('        Best parameters:', gridcvs[name].best_params_)

        # perf on test fold (valid_idx)
        outer_scores.append(gridcvs[name].best_estimator_.score(Xtr.iloc[valid_idx], Ytr.iloc[valid_idx]))
        print('        ACC (on outer test fold) %.2f%%' % (outer_scores[-1]*100))



    print('\n    Outer Loop:')
    print('        ACC %.2f%% +/- %.2f' % 
             (np.mean(outer_scores) * 100, np.std(outer_scores) * 100))
    print(f'        Total calculation time: {time() - calculation_start_time}')

# SAVE THE MODELS

gcv_model_select_DT =  gridcvs['DTreeCls']
gcv_model_select_RF =  gridcvs['RFCls']
gcv_model_select_MLP =  gridcvs['MLP']

# import joblib library
import joblib

# save Decision tree model
with open ('General_DTree_10','wb') as f:
    joblib.dump(gcv_model_select_DT,f)

# save Random forest model
with open ('General_RF_10','wb') as f:
    joblib.dump(gcv_model_select_RF,f)

# save MLP model
with open ('General_MLP_10','wb') as f:
    joblib.dump(gcv_model_select_MLP,f)
```
